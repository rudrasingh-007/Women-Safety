### Creating a function to generalize this and predict any crimes just by passing the parameters.

```python
def train_and_predict_crime(df, target_column, custom_input):
    # Step 1: Separate features and target
    features = df.columns.tolist()
    features.remove(target_column)
    features.remove('State')  # Remove non-numeric feature if it exists
    if 'Unnamed: 0' in features:
        features.remove('Unnamed: 0')  # Drop index column if it exists

    X = df[features]
    y = df[target_column]

    # Step 2: Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42
    )

    # Step 3: Train Random Forest
    model = GradientBoostingRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Step 4: Evaluate
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"Target: {target_column}")
    print(f"Mean Absolute Error: {mae:.2f}")
    print(f"R² Score: {r2:.4f}")

    # Step 5: Predict custom input
    prediction = model.predict([custom_input])
    print(f"Predicted '{target_column}' for input {custom_input}: {int(prediction[0])}")

# Driver Code
sample_input = [2025, 1000, 150, 3500, 200, 5000, 10]

# Call the function with your DataFrame and target column
train_and_predict_crime(
    df=crimes_df_cleaned,
    target_column='Rape Cases',
    custom_input=sample_input
)
```

## Creating a model for Multiple Output 

```python
from sklearn.multioutput import MultiOutputRegressor

def train_multioutput_model(df, input_features, model_type='gbr'):
    """
    Trains a multi-output regression model based on user-specified input features.

    Parameters:
    - df: DataFrame containing the dataset
    - input_features: list of column names to be used as input features
    - model_type: 'gbr' (default) for GradientBoostingRegressor

    Returns:
    - trained model
    - X_test, y_test, y_pred
    """

    # Automatically detect output columns as those not in input_features and are numeric
    output_features = [col for col in df.columns if col not in input_features and pd.api.types.is_numeric_dtype(df[col])]

    # Prepare input and output
    X = df[input_features]
    y = df[output_features]

    # Split the dataset
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Choose model
    if model_type == 'gbr':
        base_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
    else:
        raise ValueError("Currently only 'gbr' (GradientBoostingRegressor) is supported.")

    # Wrap in MultiOutputRegressor
    multi_model = MultiOutputRegressor(base_model)
    multi_model.fit(X_train, y_train)

    # Predict
    y_pred = multi_model.predict(X_test)

    # Evaluate
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred, multioutput='uniform_average')

    print(f"Model trained using inputs: {input_features}")
    print(f"Mean Absolute Error: {mae:.2f}")
    print(f"R² Score: {r2:.2f}")
    print(f"Outputs predicted: {output_features}")

    return multi_model, X_test, y_test, y_pred, output_features


# Call the function with chosen input features
input_cols = ['Year', 'Kidnap and Assault', 'Assault on Women', 'Rape Cases']
model, X_test, y_test, y_pred, output_cols = train_multioutput_model(crimes_df_cleaned, input_cols)
#train_multioutput_model(crimes_df_cleaned, in)

# Example new input (must match order of input_cols)
new_input = [[2025, 1000, 3000, 1200]]
predicted_outputs = model.predict(new_input)

# Print results
for col, val in zip(output_cols, predicted_outputs[0]):
    print(f"{col}: {int(val)}")
```
